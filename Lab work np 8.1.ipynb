```python
# Q12 Task 12: Generate a 3x3 matrix with values ranging from 1 to 9. Retrieve the element at the second row and third column.
import numpy as np
val= np.arange(1,10).reshape(3,3)
print(val)

element=val[1,2]
print(element)
```

```python
# Q13 Task 13: Slice a 4x4 array to extract the last two rows and the last two columns.

import numpy as np

data = np.arange(1,17).reshape(4,4)
print(data)

elements = data[-2,-2]
print(elements)
```

```python
# Q14 Task 14: Create a 5x5 matrix of random integers between 10 and 50. Replace all elements in the first row with the value 100.
import numpy as np
obj = np.random.randint(10,50,size=(5,5))
print(obj)

obj[0]=100
print("\nAfter replacing first row at 100: ")
print(obj)
```

```python
# Q15 Task 15: Create a 1D NumPy array of 10 elements. Use boolean indexing to filter out all values greater than 5.

arr = np.arange(1,11)
print(arr)

filter = arr[arr <= 5]
print("filtered value: ",filter)
```
